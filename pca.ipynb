```python
# Requires: tensorflow, sklearn, numpy, pandas
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, Model
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import warnings

# Suppress Keras warnings during loop training for cleaner output
warnings.filterwarnings('ignore', category=UserWarning)

# Ensure reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# -----------------------
# Example data placeholders (Assuming 'df' is loaded)
# -----------------------
# Example DataFrame creation for demonstration purposes (DELETE THIS IN YOUR SCRIPT)
#data = np.random.randn(500, 19)
columns = ['close'] + ["PB"+str(i)+"_T3" for i in range(1, 19)]
#df = pd.DataFrame(data, columns=columns)
#df['Price'] = df['Price'] * 10 + 50

# Use your specific columns
X = df[columns].dropna().values
y = X[:, 1]
X = X[:, 1:]
# Preprocess
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   # shape (N, input_dim)
N, input_dim = X_scaled.shape

# -----------------------
# Function to build and train the Autoencoder dynamically
# -----------------------
def build_and_train_autoencoder(input_dim, latent_dim_val, X_scaled_data):
    # Clear session to prevent model accumulation in memory during loop
    tf.keras.backend.clear_session()

    # Encoder
    enc_in = layers.Input(shape=(input_dim,), name="ae_input")
    h = layers.Dense(64, activation="relu")(enc_in)
    h = layers.Dense(32, activation="relu")(h)
    z = layers.Dense(latent_dim_val, activation=None, name="latent")(h)
    encoder = Model(enc_in, z, name="encoder")

    # Decoder
    dec_in = layers.Input(shape=(latent_dim_val,), name="latent_input")
    d = layers.Dense(32, activation="relu")(dec_in)
    d = layers.Dense(64, activation="relu")(d)
    recon_out = layers.Dense(input_dim, activation=None, name="reconstruction")(d)
    decoder = Model(dec_in, recon_out, name="decoder")

    # Full model
    reconstruction = decoder(encoder(enc_in))
    autoencoder = Model(enc_in, [reconstruction], name="autoencoder")

    autoencoder.compile(optimizer="adam", loss=["mse"])

    autoencoder.fit(
        X_scaled_data, [X_scaled_data],
        epochs=50,
        batch_size=32,
        verbose=1 # Silent training
    )
    return autoencoder, encoder

# -----------------------
# Loop through latent dimensions from 1 to 10 and compare
# -----------------------

comparison_results = []
max_components_to_test = 4

print(f"Starting comparison loop (1 to {max_components_to_test} components)...")

for ld in range(max_components_to_test, max_components_to_test + 1):
    # Train AE for current latent dimension
    autoencoder, encoder = build_and_train_autoencoder(input_dim, ld, X_scaled)

    # --- Calculate AE Explained Variance ---
    X_recon_full = autoencoder.predict(X_scaled, verbose=0)
    mse_reconstruction = np.mean(np.square(X_scaled - X_recon_full))
    total_original_variance_per_feature = np.var(X_scaled, axis=(0), ddof=0).mean()
    unexplained_variance_ratio = mse_reconstruction / total_original_variance_per_feature
    total_explained_variance_ratio_AE = 1.0 - unexplained_variance_ratio

    # --- Calculate PCA Explained Variance for comparison ---
    pca = PCA(n_components=ld)
    pca.fit(X_scaled)
    total_explained_variance_ratio_PCA = np.sum(pca.explained_variance_ratio_)

    # Store results
    comparison_results.append({
        'Components': ld,
        'AE_Expl_Var_Ratio': total_explained_variance_ratio_AE,
        'PCA_Expl_Var_Ratio': total_explained_variance_ratio_PCA
    })

    print(f"Components={ld}: AE Ratio={total_explained_variance_ratio_AE:.4f}, PCA Ratio={total_explained_variance_ratio_PCA:.4f}")

# -----------------------
# Display Results Summary
# -----------------------

results_df = pd.DataFrame(comparison_results)
results_df['AE_vs_PCA_Diff'] = results_df['AE_Expl_Var_Ratio'] - results_df['PCA_Expl_Var_Ratio']

print("\n" + "=" * 60)
print("Dimensionality Reduction Comparison (AE vs PCA):")
print("=" * 60)
print(results_df.to_markdown(index=False, floatfmt=".4f"))
print("=" * 60)
```
